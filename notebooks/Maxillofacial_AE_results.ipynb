```python
def plot_coverage_vs_width(
    mean_interval_widths: List[float], 
    empirical_coverages : List[float], 
    model_types: List[str],
    colors: List[str],
    markers: List[str],
    confidence_level : float = None,
    figsize: Tuple[float, float] = (8, 6)  # Nuevo argumento
) -> None:
    """
    Dibuja un scatterplot de Empirical Coverage vs. Mean Interval Width,
    coloreando los puntos según el tipo de modelo.
    
    Args:
        mean_interval_widths (list or array): valores para el eje X
        empirical_coverages (list or array): valores para el eje Y
        model_types (list or array): etiquetas de modelo para cada punto
        color_map (dict): mapa {modelo: color}
        confidence_level (int): Línea horizontal de referencia para la cobertura deseada. Por defecto: None
        figsize (tuple): Tamaño de la figura en pulgadas, como (ancho, alto). Por defecto: (10,8)
    """
    
    if not (len(mean_interval_widths) == len(empirical_coverages) == len(model_types) == len(colors)):
        raise ValueError("All input lists must have the same length.")
    
    plt.figure(figsize=figsize)
    plt.grid(True, zorder=0)
    
    # Agrupar por tipo de marcador para graficarlos por separado
    for marker in set(markers):
        idx = [i for i, m in enumerate(markers) if m == marker]
        plt.scatter(
            [mean_interval_widths[i] for i in idx],
            [empirical_coverages[i] for i in idx],
            c=[colors[i] for i in idx],
            marker=marker,
            s=100,
            edgecolor='black',
            linewidth=0.5,
            zorder=3
        )
    
    # Línea horizontal en y=confidence_level
    if confidence_level is not None:
        plt.axhline(y=confidence_level, color='black', linestyle='--', linewidth=1.5, zorder=2)

    # Etiquetas y título
    plt.xlabel("Mean Interval Width")
    plt.ylabel("Empirical Coverage")
    plt.title("Empirical Coverage vs. Mean Interval Width")

    # Legend
    seen = set()
    legend_elements = []
    for model, color in zip(model_types, colors):
        if model not in seen:
            legend_elements.append(
                plt.Line2D([0], [0], marker='o', color='w', label=model,
                           markerfacecolor=color, markersize=10)
            )
            seen.add(model)

    plt.legend(handles=legend_elements, title="Model Type")
    plt.tight_layout()
    plt.show()
```

```python
# Define la clase MaxillofacialXRayDataset, que se utiliza para cargar imágenes de rayos X maxilofaciales 
# junto con su edad correspondiente desde un fichero de metadatos. 
class MaxillofacialXRayDataset(Dataset):
    
    def __init__(self, metadata_file, images_dir, transform=None):
        """
        metadata_file: Ruta al fichero de metadatos (CSV u otro formato)
        images_dir: Ruta al directorio de imágenes (entrenamiento o prueba)
        transform: Transformaciones a aplicar a las imágenes (normalización, etc.)
        """
        metadata = pd.read_csv(metadata_file)  
        self.images_dir = images_dir
        self.transform = transform
        
        # Preprocesar los campos una sola vez
        self.img_paths = metadata['ID'].apply(lambda id_: os.path.join(images_dir, id_)).tolist()
        self.sexes = torch.tensor((metadata['Sex'] != 'M').astype(int).values, dtype=torch.long)
        self.ages = torch.tensor(metadata['Age'].values, dtype=torch.float32)
    
    
    def __len__(self):
        return len(self.img_paths)
    
    
    def __getitem__(self, idx):
        
        # Carga la imagen
        image = Image.open(self.img_paths[idx]).convert('RGB')
        if self.transform:
            image = self.transform(image)

        return image, self.sexes[idx], self.ages[idx]
```

```python
test_transform = transforms.Compose(
    [transforms.Resize((448, 224)),
     transforms.ToTensor(),
     transforms.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))]
)

# Crea el Dataset de test con solo resize y normalización
testset  =  MaxillofacialXRayDataset(
    metadata_file = data_dir + 'metadata_test.csv',
    images_dir = data_dir + 'test/',
    transform = test_transform
) 
```

```python
# Establece un batch size de 32 
BATCH_SIZE = 32

# Crea DataLoader de test
test_loader = DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
df = pd.DataFrame(columns=[
    "pred_model_type", 
    "sex_embedding",
    "confidence", 
    "pred_point_values",
    "pred_lower_bound",
    "pred_upper_bound",
    "true_values"
])

MODEL_CLASSES = {
    'base': ResNeXtRegressor,
    'QR': ResNeXtRegressor_QR,
    'ICP': ResNeXtRegressor_ICP,
    'CQR': ResNeXtRegressor_CQR,
    'CRF': ResNeXtRegressor_CRF
}

load_model_paths = ['models/AE_model01_base.pth', 'models/AEsex_model01_base.pth',
                    'models/AE_model02_ICP.pth', 'models/AEsex_model02_ICP.pth',
                    'models/AE_model03_QR.pth', 'models/AEsex_model03_QR.pth',
                    'models/AE_model04_CQR.pth', 'models/AEsex_model04_CQR.pth']


for load_model_path in load_model_paths:
    
    #
    checkpoint = torch.load(load_model_path, weights_only=False)
    pred_model_type = checkpoint['pred_model_type']
    use_metadata = checkpoint['use_metadata']
    confidence = 1-checkpoint['alpha'] if 'alpha' in checkpoint else None

    # Obtiene la clase del modelo basado en el checkpoint
    model_class = MODEL_CLASSES.get(checkpoint['pred_model_type'])

    # Crea el modelo
    model = model_class(
        confidence = confidence, 
        use_metadata = use_metadata, 
        meta_input_size = 1 if use_metadata else 0
    ).to('cuda')

    # Carga el modelo
    model.load_checkpoint(checkpoint)

    if pred_model_type == 'base':
        test_pred_point_values, test_pred_lower_bound, test_pred_upper_bound, test_true_values = \
            model.inference(test_loader, test_loader)                                                         # A corregir
    
    else:
        test_pred_point_values, test_pred_lower_bound, test_pred_upper_bound, test_true_values = \
            model.inference(test_loader)
            
    # Convierte todos los arrays a CPU y NumPy si son tensores
    test_pred_point_values = test_pred_point_values.cpu().numpy()
    test_pred_lower_bound = test_pred_lower_bound.cpu().numpy()
    test_pred_upper_bound = test_pred_upper_bound.cpu().numpy()
    test_true_values = test_true_values.cpu().numpy()

    interval_widths = test_pred_upper_bound - test_pred_lower_bound
    
    
    # Agrega cada predicción individual como una fila en el DataFrame
    for i in range(len(test_pred_point_values)):
        df.loc[len(df)] = [
            pred_model_type,
            use_metadata,
            confidence,
            test_pred_point_values[i],
            test_pred_lower_bound[i],
            test_pred_upper_bound[i],
            test_true_values[i]
        ]
    
# Crea la columna interval_width
df["interval_width"] = df["pred_upper_bound"] - df["pred_lower_bound"]
```

```python
def plot_confidence_predictions(
    pred_point_values : torch.Tensor,
    pred_lower_bound : torch.Tensor,
    pred_upper_bound : torch.Tensor,
    true_values : torch.Tensor
) -> None:
    
    # Convierte los tensores a NumPy
    y_true = true_values.numpy()
    y_pred = pred_point_values.numpy()
    y_lower = pred_lower_bound.numpy()
    y_upper = pred_upper_bound.numpy()
    
    # Ordena por los valores predichos
    order = np.argsort(y_true)
    y_true = y_true[order]
    y_pred = y_pred[order]
    y_lower = y_lower[order]
    y_upper = y_upper[order]
    
    x = np.arange(len(y_true))
    
    # Grafica
    plt.figure(figsize=(12, 6))
    plt.plot(x, y_true, label='Valor verdadero', 
             color='black', linewidth=2, marker='s', markersize=4)
    plt.plot(x, y_pred, label='Predicción puntual', 
             color='steelblue', linestyle='--', marker='o', markersize=4)
    plt.fill_between(x, y_lower, y_upper, 
                     color='lightcoral', alpha=0.5, label='Intervalo de confianza')

    plt.xlabel('Instancia (ordenada por valor verdadero)')
    plt.ylabel('Valor')
    plt.title('Predicción con intervalos de confianza')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_for_model_and_sex(df, model_type, sex_embedding, n=100):
    # Filtrar DataFrame por modelo y sex_embedding
    df_filtered = df[(df['pred_model_type'] == model_type) & (df['sex_embedding'] == sex_embedding)]
    
    # Extraer los valores y convertir a tensores
    pred_point_values = torch.tensor(df_filtered['pred_point_values'].values[0:n])
    pred_lower_bound = torch.tensor(df_filtered['pred_lower_bound'].values[0:n])
    pred_upper_bound = torch.tensor(df_filtered['pred_upper_bound'].values[0:n])
    true_values = torch.tensor(df_filtered['true_values'].values[0:n])
    
    # Llamar a la función de plot
    plot_confidence_predictions(
        pred_point_values=pred_point_values,
        pred_lower_bound=pred_lower_bound,
        pred_upper_bound=pred_upper_bound,
        true_values=true_values
    )
```
